#### Logs

##### Security logs

This shows successful logins and other logs related to security. Probably priviledge escalation as well.


```powershell
Get-WinEvent -FilterHashtable @{Logname='security';}  -max 5  | select TimeCreated,ID,Message | ConvertTo-CSV
```

##### application logs

This shows my attempts to install *Log Parser* (which was unsuccessful). If this was more than 5 entries, it would show more logs produced by different applicatios.

```powershell
Get-WinEvent -FilterHashtable @{Logname='application';}  -max 5  | select TimeCreated,ID,Message | ConvertTo-CSV
```

##### system logs

This is showing changes to the registry 

```powershell
Get-WinEvent -FilterHashtable @{Logname='system';}  -max 5  | select TimeCreated,ID,Message | ConvertTo-CSV
```

##### File Listings

This shows the files in C:\

```powershell
gci -Path $Env:systemdrive\ -force -ea 0 | select fullname,CreationTime,lastwritetime | ConvertTo-CSV 
```

I wanted to check how to list files in a different directory, the change to the 
above command was super simple; just add the directory path.

```powershell
gci -Path "C:\Program Files" -force -ea 0 | select fullname,CreationTime,lastwritetime | ConvertTo-CSV 
```
